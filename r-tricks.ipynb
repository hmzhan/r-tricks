### R trick: #1: Keep rows having smallest values in each group

```r
library(dplyr)
starwars %>% group_by(sex) %>% slice_min(height, with_ties=FALSE)
```

```r
starwars %>% group_by(sex) %>% slice(which.min(height))
```

```r
starwars %>% group_by(sex) %>% filter(height == min(height, na.rm=TRUE))
```

### R trick #2: Filter groups with group summaries

```r
starwars %>% group_by(sex) %>% summarize(count = n())
```

```r
starwars %>% group_by(sex) %>% filter(n() > 10)
```

```r
starwars %>% group_by(sex) %>% summarize(avg_height = mean(height, na.rm=TRUE))
```

```r
starwars %>% group_by(sex) %>% filter(mean(height, na.rm=TRUE) > 170)
```
